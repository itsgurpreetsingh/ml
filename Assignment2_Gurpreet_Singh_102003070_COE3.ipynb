Merging two datasets and assigning to data

```python
data=pd.merge(customerdata,sales,on='CustomerID')
data.head()
```

Dropping unnecessary attributes.

```python
data=data.drop(['FirstName','LastName','MiddleName','CustomerID','NumberCarsOwned','NumberChildrenAtHome','TotalChildren'],axis=1)
```

```python
data.info()
```

Part 2

As NULL values are of no use so we can remove Title,Suffix,AddressLine2 from data set

```python
data=data.drop(['Title','Suffix','AddressLine2'],axis=1)
```

```python
data
```

Normalization of YearlyIncome

```python
df_min_max_scaled = data.copy()
for column in df_min_max_scaled.columns:
    if column=="YearlyIncome":
        df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  
print(df_min_max_scaled)
```

```python
from scipy import stats
df = pd.DataFrame(data)
df['YearlyIncome'] = stats.zscore(df['YearlyIncome'])
print(df)
```

One Hot Encoding

```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['Education','Occupation','MaritalStatus','HomeOwnerFlag','Gender'])
print(one_hot_encoded_data)
```

Part 3

Cosine Similarity

```python
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial import distance
df['Education'] = pd.factorize(df['Education'])[0]
scaler=MinMaxScaler()
scaled=scaler.fit_transform(df[['YearlyIncome','AvgMonthSpend']])
df['YearlyIncome_scaled']=scaled[:,0]
df['AvgMonthSpend_scaled']=scaled[:,1]
df.drop(['YearlyIncome','AvgMonthSpend'],axis=1,inplace=True)
distance.cosine(df['Education'],df['AvgMonthSpend_scaled'])
```

Jaccard Similarity

```python
distance.jaccard(df['Education'].values,df['AvgMonthSpend_scaled'].values)
```
